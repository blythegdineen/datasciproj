## Data Description

We are using four different datasets.

This report seems to have been made to sell to companies to find reference points within industries to compare to competitors. As mentioned in our limitations, the companies involved in this report are all voluntary, so data is self-reported and therefore incomplete, with room for inaccuracies. We retrieved this data from the Statista website. No people were involved in the survey and all data was voluntarily given. The numbers come from data on government filings and published financial reports. Single industry values, like the cigarette industry, are calculated using only the companies that participate, which varies from year to year. https://drive.google.com/file/d/1FTkqcqQoo6ekZE4gfwVKp4qAZn9PXnql/view?usp=sharing

It is important to note that for the two datasets previously described the rows with * next to the year are predicted amounts and not actually records. 

The third is “Tobacco production in the US 2000-2019”, which compares Years and Tobacco production in the United States(in 1,000 pounds). This was a fairly objective count of tobacco production by producers in the United States, presumably some of the numbers were rounded because producers are not weighing all of their stock but instead giving an accurate estimate. This also does not include people who produced tobacco for personal consumption, instead only including nationally recorded producers, however that quantity is likely so limited it is negligible. https://drive.google.com/file/d/1dEbS4yPHkGaO708Rrg7-yJP46DIHG13n/view?usp=sharing

### Summary Functions

For the first part of our data analysis, we wanted to get the mean, median, max, and min for each of our datasets. This gives us a good start of understanding the data and seeing trends by looking at the years of min and max. As a note, this is all production and consumption in the United States.

```python
mean = tproduction['Tobacco Production'].mean()
median = tproduction['Tobacco Production'].median()
max = tproduction['Tobacco Production'].max()
index = tproduction[tproduction['Tobacco Production']==max].index.values[0]
maxyear = tproduction.loc[index,'Year']
min = tproduction['Tobacco Production'].min()
index2 = tproduction[tproduction['Tobacco Production']==min].index.values[0]
minyear = tproduction.loc[index2,'Year']
print("The mean tobacco production from 2000 to 2019 is {:.0f} pounds.".format(mean))
print("The median tobacco production from 2000 to 2019 is {:.0f} pounds.".format(median))
print("The max tobacco production from 2000 to 2019 is {:.0f} pounds.".format(max))
print("The max occured in " + str(maxyear) +".")
print("The min tobacco production from 2000 to 2019 is {:.0f} pounds.".format(min))
print("The min occured in " + str(minyear) + ".")
```

The minimum production occuring in the first year of the dataset and the maximum production occuring in the last year of the dataset goes with our hypothesis that tobacco production has been declining. 

```python
mean = ads18["US Cigarettes Ad Spending (in millions)"].mean()
median = ads18["US Cigarettes Ad Spending (in millions)"].median()
max = ads18["US Cigarettes Ad Spending (in millions)"].max()
index = ads18[ads18["US Cigarettes Ad Spending (in millions)"]==max].index.values[0]
maxyear = ads18.loc[index,'Year']
min = ads18["US Cigarettes Ad Spending (in millions)"].min()
index2 = ads18[ads18["US Cigarettes Ad Spending (in millions)"]==min].index.values[0]
minyear = ads18.loc[index2,'Year']
print("The mean spent on cigarette advertisting from 2018 to 2020 is {:.0f} millions of dollars.".format(mean))
print("The median spent on cigarette advertisting from 2018 to 2020 is {:.0f} millions of dollars.".format(median))
print("The max spent on cigarette advertisting from 2018 to 2020 is {:.0f} millions of dollars.".format(max))
print("The max occured in " + str(maxyear) + '.')
print("The min spent on cigarette advertisting from 2018 to 2020 is {:.0f} millions of dollars".format(min))
print("The max occured in " + str(minyear) + '.')
```

The third analysis we wanted to do was to complete a clustering by type of tobacco product compared to consumption and analyze trends in types. We did this by looking at Year, Total consumption, and separating it by type. We then created 5 clusters (because there are 5 types of tobacco: smokeless tobacco, cigarettes, cigars, loose tobacco, and all combustibles (aka total)). We used these clusters to try and predict the type based on the total consumption for each year. However, as seen in the low accuracy, this clustering was not too effective at predicting the actual type. 

```python
clusterconsum = consum.loc[:,["Year", "Total", "Type"]].copy()
sns.relplot(x="Year", y="Total", hue="Type", data=clusterconsum)
plt.xlabel("Year")
plt.ylabel("Total")
plt.title("Total Tobacco Consumption based on Type")
plt.show()
```

This graph gives a visual to the consumption based on type, by showing the types in different colors. Cigarettes and all combustibles are much higher and much more seperated than the other three types which all sort of overlap each other. 

```python
clustering = KMeans(n_clusters = 5, random_state = 15)
clustering.fit(clusterconsum[["Year", "Total"]])
cluster_labels = clustering.labels_
clusterconsum["cluster_label"] = cluster_labels
predicted_type_lookup = pd.DataFrame({"cluster_label": [0,1,2,3,4], "predicted_type": ["Smokeless Tobacco", "Cigarettes", "Cigars", "Loose Tobacco", "All Combustibles"]})
type_predict = pd.merge(clusterconsum, predicted_type_lookup)
accurate = 0 
for i in range(len(type_predict)):
    if (type_predict.loc[i]["Type"]==type_predict.loc[i]["predicted_type"]):
        accurate += 1
print("The percent of accurate predictions is {:.1f}%".format(100*(accurate/len(type_predict))))
```

```python
sns.relplot(x="Year", y="Total", hue="Type", style="cluster_label", data=clusterconsum)
plt.xlabel("Year")
plt.ylabel("Total")
plt.title("Total Tobacco Consumption based on Type and Clustering")
plt.show()
```

In this graph, we added the style of the data point to be the prediction based on the clustering. This shows the lack of accuracy. If our clustering was better at predicting the type, each color would mostly correspond to one shape, which is not the case.

## Limitations

Limitations for each data set:
- For the data on cigarette advertising from 2011 to 2015 there are a few important limitations to be noted. First off, the data for 2015 is a projected value. The survey was form 2011 to 2014, so the 2015 data point is just a forecast. Secondly, the data only includes companies that participated in the survey. This creates a potential for a lack of trust in this data. We don't know how Schonfeld & Associates conducted the survery or how they got companies to participate. It's possible they were only looking at a particular set of companies or that the companies that did agree do not represent the entire cigarette industry. It mentions that the companies that participate varies from year to year which adds another asepct of variability. This also makes the 2015 data point extra unreliable because they didn't know at the time which companies were going to participate in 2015.

- Because the data on cigarette advertising from 2018 to 2020 comes from the same Schonfeld & Associates report, it has all of the same limitations. In this case, 2019 and 2020 are predicted. Since there are only three datapoints, having two of them be forecast values, and not actual values, definitely puts an extra limitation on what we can really use this dataset for. As explained earlier, the companies differ year to year, so any trend we see in this data may not be completely reliable.

- As for the production data set, since it is just comparing US production to the year, it presents crop yield. However, this does not factor in the market climate (say, the stock market crash of 2008 may have had an effect on tobacco producers) or a bad climate year (weather or poor crop yield as opposed to lack of demand forcing down production levels) or imported tobacco products, only US produced products, while China is the world's largest tobacco producer, so an increase in cheaper imports may drive down production of domestic tobacco.

- With the consumption data set, it is not clear what tobacco products are e-cigarettes, since their popularity really exploded over the past five years and this data set starts 15 years before that boom, therefore the tobacco products are more focused on cigarettes, cigars, and loose tobacco instead of vapes and e-cigarettes. Additionally, the quantities are given in inconsistent units so while converting cigarettes and cigars to pounds of tobaccco loses some of the specificity of the data, otherwise the data would be incomparable.

General limitations:
What would have been really interesting is to look into the advent of e-cigarettes but that could have been a project all of its own.
